```python
import numpy as np

print("1. np.arange() - Counting with steps:")
arr1 = np.arange(1, 11)           # start=1, stop=11 (11 not included)
print(arr1)

arr2 = np.arange(0, 20, 5)        # step = 5
print("\nWith step 5:", arr2)

print("\n2. np.linspace() - Evenly spaced numbers:")
arr3 = np.linspace(0, 10, 5)      # 5 numbers from 0 to 10
print(arr3)

print("\n3. np.logspace() - Logarithmic scale:")
arr4 = np.logspace(0, 3, 4)       # 10^0 to 10^3 → 4 numbers
print(arr4)
```

```python
### 🔍 Iterating Using Loops

The simplest way to iterate over a NumPy array is by using Python loops.

For a 1D array, iteration is straightforward — each element is accessed one by one using a single loop.

For multi-dimensional arrays:
- A single loop iterates over rows
- Nested loops are required to access each individual element

This method is easy to understand and useful for beginners. However, it is not efficient for large datasets because Python loops are slower compared to NumPy’s optimized operations.


### ⚙️ Using np.nditer()

NumPy provides a built-in iterator called `np.nditer()` for efficient array traversal.

It allows iteration over arrays of any dimension without writing multiple nested loops.

Key points:
- Works for multi-dimensional arrays
- More efficient than standard Python loops
- Provides flexibility in how elements are accessed

This makes `np.nditer()` a preferred choice when dealing with large arrays or performance-critical tasks.


### 🔄 Iterating with Different Orders

NumPy arrays can be iterated in different memory orders, which affects how elements are accessed.

There are two main types:

- **C-order (Row-major order):**
  Elements are stored row by row. Iteration happens across rows first. This is the default in NumPy.

- **F-order (Column-major order):**
  Elements are stored column by column. Iteration happens down columns first.

Understanding iteration order is important because:
- It affects performance
- It determines how data is accessed internally
- It becomes important in large-scale computations

NumPy allows control over iteration order, especially when using functions like `np.nditer()`.
```

```python
import numpy as np

arr=np.array([

              [1,2],
              [3,4]
])

print('Using loop')

for row in arr:
  for element in row:

    print(element)


for x in np.nditer(arr):
  print(x)
```

```python
import numpy as np

marks=np.array([

              [78,85,92],
              [65,88,71],
              [95,82,79]
])


count=0

for x in np.nditer(marks,order='F'):
  if x>=80:
    count+=1
    print(x)


print(count)






```

```python
import numpy as np

marks = np.array([[78, 85, 92],
                  [65, 88, 71],
                  [95, 82, 79]])

print("Student Marks:")
print(marks)

print("\n--- Using np.nditer() with index ---")
count_above_80 = 0

for x in np.nditer(marks):
    print(x, end=" ")
    if x > 80:
        count_above_80 += 1

print("\n\nNumber of students who scored above 80:", count_above_80)

print("\n--- Iterating with different order (Column-wise) ---")
print("Column-major order (F order):")
for x in np.nditer(marks, order='F'):
    print(x, end=" ")
```

```python
🔹 TOPIC: Binary Operators in NumPy

👉 Bitwise operations
👉 Logical operations

1. THEORY (clear + simple)
✅ What are Binary Operators?

👉 Operations that work on two values (element-wise in arrays)


🧠 Real-world analogy

Think:

You compare marks of students
Or combine conditions

👉 Example:

Marks > 50
Marks < 90
```

```python
import numpy as np

a = np.array([0, 1, 2, 3, 4])
b = np.array([5, 4, 3, 2, 1])

print("a =", a)
print("b =", b)

print("\n--- Bitwise Operations ---")
print("a & b (AND):", a & b)
print("a | b (OR): ", a | b)
print("a ^ b (XOR):", a ^ b)
print("~a (NOT):   ", ~a)

print("\n--- Logical Operations ---")
print("Logical AND:", np.logical_and(a > 2, b < 4))
print("Logical OR: ", np.logical_or(a > 2, b < 4))
```

```python
import numpy as np

# Columns: [Math, Science, English]
scores = np.array([[85, 78, 65],
                   [92, 88, 95],
                   [45, 70, 80],
                   [78, 65, 72],
                   [95, 90, 85]])

math = scores[:, 0]
science = scores[:, 1]

print("Math scores:", math)
print("Science scores:", science)

# Using Logical Operations (Recommended for conditions)
good_in_both = np.logical_and(math >= 80, science >= 80)
good_in_one = np.logical_or(math >= 90, science >= 90)

print("\nStudents good in both Math & Science:", good_in_both)
print("Students good in at least one (90+):", good_in_one)


# Using Bitwise operators (very common in NumPy)
excellent = (math >= 85) & (science >= 85)          # AND
outstanding = (math >= 90) | (science >= 90)        # OR

print("\nExcellent students (using &):", excellent)
print("Outstanding students (using |):", outstanding)

# Get actual student scores using boolean indexing
print("\nExcellent students' scores:\n", scores[excellent])
```
