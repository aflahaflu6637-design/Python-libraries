```python
🔍 Advanced Indexing — Boolean Indexing in Pandas

📌 Week 1 Connection: Remember if statements and conditions like age > 18? And from earlier today — Binary Operators like &, |, ~? Boolean Indexing is exactly those two things applied to entire tables!


1. 📚 THEORY
What is it and why do we use it?
Boolean Indexing means filtering rows from a DataFrame by asking a True/False question about each row.
Pandas checks every row, and only keeps the rows where the answer is True.
Without Boolean Indexing — manual and painful 😫
for each student:
    if score > 80:
        print that student

With Boolean Indexing — one clean line 😍
df[df["Score"] > 80]

🌍 Real World: This is exactly how apps filter data —
"Show me only flights under ₹5000"
"Show me only products in stock"
"Show me only customers from Kerala"


🔑 Key Concepts — Real World Analogy
🚦 Think of Boolean Indexing as a Security Checkpoint
Imagine every row in your DataFrame is a person walking through a checkpoint.
The condition is the security rule:
Rule: "Only let through people whose age > 18"

Row 0 → Alice, age 20 → ✅ PASS (True)  → kept
Row 1 → Bob,   age 16 → ❌ FAIL (False) → removed
Row 2 → Carol, age 22 → ✅ PASS (True)  → kept
Only the rows that pass the check (True) make it into your result!
```

```python
import pandas as pd

# Our DataFrame for all examples
data = {
    "Name":    ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Subject": ["Math", "Science", "English", "Math", "Science"],
    "Score":   [85, 92, 78, 95, 88],
    "Age":     [20, 22, 19, 21, 23],
    "Passed":  [True, True, False, True, True]
}

df = pd.DataFrame(data)
print("=== Full DataFrame ===")
print(df)


print("\n--- Students with Score > 80 ---")
print(df[df['Score']>80])

print("\n--- Students who study Math ---")
print(df[df['Subject']=='Math'])

print("\n--- Students who did NOT pass ---")
print(df[df['Passed']==False])


print("\n--- Students who passed (cleaner way) ---")
print(df[df['Passed']])

print("\n--- Students aged 20 or above ---")

print(df[df['Age']>=20])
```

```python
import pandas as pd

# Our DataFrame for all examples
data = {
    "Name":    ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Subject": ["Math", "Science", "English", "Math", "Science"],
    "Score":   [85, 92, 78, 95, 88],
    "Age":     [20, 22, 19, 21, 23],
    "Passed":  [True, True, False, True, True]
}

df = pd.DataFrame(data)
print("=== Full DataFrame ===")
print(df)


# & = AND (both must be True)
print("\n--- Score > 80 AND Age < 22 ---")

print(df[(df['Score']>80) & (df['Age']<22)])


# | = OR (at least one must be True)
print("\n--- Score > 90 OR Subject is Math ---")

print(df[(df['Score']>90) | (df['Subject']=='Math')])


# ~ = NOT (flip the condition)
print("\n--- NOT Science students ---")


print(df[df['Subject']!='Science'])
```

```python
TOPIC: Data Transformation in Pandas

👉 groupby()
👉 pivot_table()

1. THEORY (clear + simple)
✅ What is Data Transformation?

👉 Changing raw data into a useful summarized form

🧠 Real-world analogy

Imagine you are a teacher:

You have marks of 100 students
You want:
average marks per class
total marks per subject

👉 You group students and calculate results
```

```python
data = {
    "Class": ["A", "A", "B", "B","A","A","B","B"],
    "Subject": ["Math", "Science","Math", "Science","Science","Science","Math","Math"],
    "Marks": [80, 90, 70, 60,70,80,70,50]
}

df = pd.DataFrame(data)
print(df)

print(df.groupby(['Class','Subject'])['Marks'].agg(['sum','mean']))
```

```python
data = {
    "City": ["Delhi", "Delhi", "Mumbai", "Mumbai", "Delhi"],
    "Product": ["A", "B", "A", "B", "A"],
    "Sales": [100, 150, 200, 250, 120]
}

df = pd.DataFrame(data)
print(df)



pivot=pd.pivot_table(
    df,
    values='Sales',
    index='City',
    aggfunc='sum',
    columns='Product'
)

print(pivot)
```

```python
import pandas as pd

data={
    'Class':['A','A','B','B'],
    'Marks':[80,90,70,60]
}


df=pd.DataFrame(data)

print(df)


print(df.groupby('Class')['Marks'].sum())

print(df.groupby('Class')['Marks'].min())

print(df.groupby('Class')['Marks'].agg(['sum','mean','min','max']))
```

```python
import pandas as pd

data = {
    "City": ["Delhi", "Delhi", "Mumbai", "Mumbai", "Delhi"],
    "Sales": [100, 150, 200, 250, 120]
}

df = pd.DataFrame(data)

# Total sales per city
total_sales = df.groupby("City")["Sales"].sum()

# Average sales per city
avg_sales = df.groupby("City")["Sales"].mean()

print("Total Sales:\n", total_sales)
print("\nAverage Sales:\n", avg_sales)
```
