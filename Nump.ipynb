```python
import numpy as np

marks=np.array([45,60,75,90,55])
updated_marks=marks+5

avg_mark=np.mean(updated_marks)

print('Ornginal marks:',marks)
print('Updated Marks: ',updated_marks)
print('Average Mark:',avg_mark)

print(marks.dtype)

print('Avrage mark:',marks.mean())
```

```python
import numpy as np

sales=np.array([[250,300,180],
                [320, 290, 210]])

print('Sales: ')
print(sales)

print('Sales per product in 2 weeks :',sales.sum(axis=0))

print('Total sales per week: ',sales.sum(axis=1))
print('Average sales per product: ',sales.mean(axis=0))
print('Average sales per week :',sales.mean(axis=1))
print(sales.shape)
print(sales.dtype)
print(sales.shape)

sales_week=sales.sum(axis=1)
print(sales_week)

total_sales=sales_week.sum(axis=0)
print(total_sales)


sales_per_item=sales.sum(axis=0)
print(sales_per_item)

total_sale=sales_per_item.sum()
print(total_sale)


print('average_sale:',sales.mean())

print(sales.max())

print(sales.max(axis=0))

print(sales.max(axis=1))
```
