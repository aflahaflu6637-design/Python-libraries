```python
import matplotlib.pyplot as plt
import datetime as dt

# time data
dates = [
    dt.datetime(2024, 1, 1),
    dt.datetime(2024, 1, 2),
    dt.datetime(2024, 1, 3),
    dt.datetime(2024, 1, 4)
]

# values
values = [10, 20, 15, 25]

plt.plot(dates, values)
plt.xticks(rotation=45)
plt.title("Time Series Plot")
plt.xlabel("Date")
plt.ylabel("Value")

plt.show()
```

```python
import matplotlib.pyplot as plt

# X values (same for all lines)
x = [1, 2, 3, 4, 5]

# 5 different Y values
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [3, 4, 5, 6, 7]
y4 = [4, 5, 6, 7, 8]
y5 = [5, 6, 7, 8, 9]

# Plot 5 lines
plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.plot(x, y3, label="Line 3")
plt.plot(x, y4, label="Line 4")
plt.plot(x, y5, label="Line 5")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("5 Lines in One Graph")

# Show legend
plt.legend()
# Show graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
subjects = ["Math", "Science", "English", "Math", "Science", "English"]
marks = [80, 75, 78, 85, 82, 88]
gender = ["Male", "Male", "Male", "Female", "Female", "Female"]

# Create point plot
sns.pointplot(x=subjects, y=marks, hue=gender)

plt.title("Average Marks by Subject and Gender")
plt.show()
```

```python
# 1. Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Create Dataset
data = {
    "hours_studied": [2, 3, 4, 5, 6, 7, 8, 9],
    "marks": [50, 55, 65, 70, 75, 80, 85, 95],
    "gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun", "Mon"]
}

df = pd.DataFrame(data)

# 3. Apply Styling
sns.set_theme(style="whitegrid")     # Theme
sns.set_palette("pastel")            # Color palette

# 4. Plot
sns.scatterplot(x="hours_studied", y="marks", hue="gender", data=df)

# 5. Customize
plt.title("Study Hours vs Marks")
plt.grid(axis='y')                   # Grid control

# 6. Show Plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "x": [1,2,3,4,5],
    "y": [10,15,20,25,30]
}
df = pd.DataFrame(data)
sns.set_palette('deep')
sns.set_style('darkgrid',{

                        'grid.color':'red',
                        'grid.linestyle':'--'
})
sns.lineplot(x="x", y="y", data=df,lw=3,color='green')



plt.title("custom grid style")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    "category": ["A","B","C","A","B","C"],
    "value": [10,20,15,25,30,22]
}


df = pd.DataFrame(data)

sns.set_theme(style="whitegrid")
sns.set_palette("coolwarm")

sns.barplot(x="category", y="value", hue="category", data=df)

plt.title("Pastel Palette")
plt.show()
```
